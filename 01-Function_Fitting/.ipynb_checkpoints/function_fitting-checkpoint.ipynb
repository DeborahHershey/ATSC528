```python
### Read in observations ###
anum = np.loadtxt("RAOBs_201903131200.txt",
                 delimiter=",",usecols=(1,2,3))
aplace = np.loadtxt("RAOBs_201903131200.txt",
                 delimiter=",",usecols=0,dtype='str')
display(anum)
print(len(anum[:,0]))
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###
x0 = 18.9
y0 = -6.3
xgrid = x0+np.arange(22)*1.27
ygrid = y0+np.arange(28)*1.27 #in southwest corner

x,y = np.meshgrid(xgrid, ygrid, indexing='xy')
#print(alist)

plt.plot(x, y, marker='o', color='k', linestyle='none')

plt.show()
```

```python
### convert obs lat/long to x,y (may want to plot on your analysis grid to verify)###
'''
Note: to convert between lat/lon, we use the general equation
 y = m*rho*sigma*cos(phi)*sin(lambda0 - lambda), where rho = avg radius of earth (in cm), phi = lat, lambda = lon, lambda0 = 60N (given), m = map conversion factor, and sigma = ((1+sin(phi0))/(1+sin(phi)))
 x = m*rho*sigma*cos(phi)*cos(lambda0 - lambda)
'''
xobs = np.empty(len(anum))
yobs = np.empty(len(anum))
m = 1/15000000
l = -115*np.pi/180
phi0 = 60
r = 637100000 #centimeters

for n in np.arange(len(anum)):
    xobs[n] = m*r*((1+np.sin(60*np.pi/180))/(1+np.sin(anum[n,0]*np.pi/180)))*np.cos(anum[n,0]*np.pi/180)*np.cos((anum[n,1]*np.pi/180)-l)
    yobs[n] = m*r*((1+np.sin(60*np.pi/180))/(1+np.sin(anum[n,0]*np.pi/180)))*np.cos(anum[n,0]*np.pi/180)*np.sin((anum[n,1]*np.pi/180)-l)


#print()
plt.plot(xobs,yobs,marker='.',linestyle='none')
plt.plot(x, y, marker='.', color='k', linestyle='none')
plt.show
```

```python
### Perform 500mb geopotential height analyses using a second order 2-d polynomial with two ###
### radii of influence (10cm & 20cm) ###

#setting up variables for this and later steps

roi10 = []
roi20 = []
lenroi10=[]
lenroi20=[]
a10 = []
a20 = []
temp = 0
#matrix analysis for any point's radius of influence
def analysis(x,y,mb):
    a = [[1],[(1/len(x))*sum(x)],[(1/len(y))*sum(y)],[(1/len(x))*sum(x**2)],[(1/len(y))*sum(y**2)],[(1/len(x))*sum(x)*(1/len(y))*sum(y)]]*[1,(1/len(x))*sum(x),(1/len(y))*sum(y),(1/len(x))*sum(x**2),(1/len(y))*sum(y**2),(1/len(x))*sum(x)*(1/len(y))*sum(y)] #6x6 xy matrix
    final = np.linalg.inv(a)*[[1/len(mb)*sum(mb)],[(1/len(x))*sum(x)*1/len(mb)*sum(mb)],[(1/len(y))*sum(y)*1/len(mb)*sum(mb)],[(1/len(x))*sum(x**2)*1/len(mb)*sum(mb)],[(1/len(y))*sum(y**2)*1/len(mb)*sum(mb)],[(1/len(x))*sum(x)*(1/len(y))*sum(y)*1/len(mb)*sum(mb)]] #multiplying inverse 6x6 matrix to 500mb obs to get coefficient matrix
    return final[0] #returning C00 coefficient value, IE Analysis weight

#for loop for 10 cm distance roi: stores indices of grid points of what obs fall within what grid points of roi
for i in np.arange(len(x[:,0])):
    for j in np.arange(len(y[0,:])):
        for k in np.arange(len(xobs)):
            if np.sqrt((xobs[k]-x[i,0])**2) + np.sqrt((yobs[k]-y[0,j])**2) <= 10:
                t,*tt,ta=anum[k]
                tb=xobs[k]
                tc=yobs[k]
                roi10.append([i,j,k,ta,tb,tc])
                temp+=1
            else:
                k+=1
        else:
            lenroi10.append([i,j,temp]) #counts how many obs fall within each analysis roi
            j+=1
            temp=0
    else:
        i+=1

'''
#for loop for 20 cm distance roi
for i in np.arange(len(x[:,0])):
    for j in np.arange(len(y[0,:])):
        for k in np.arange(len(xobs)):
            if np.sqrt((xobs[k]-x[i,0])**2) + np.sqrt((yobs[k]-y[0,j])**2) <= 20:
                roi10.append([i,j,k,anum[k,2],xobs[k],yobs[k]])
                d10.append([i,j,k,np.sqrt((xobs[k]-x[i,0])**2) + np.sqrt((yobs[k]-y[0,j])**2)])
            else:
                k+=1
        else:
            lenroi20.append([i,j,temp])
            j+=1
            temp=0
    else:
        i+=1
'''
# to plot map points back onto lat/lon map
def latlongify(x,y):
    phi = (np.pi/2) - 2*np.arctan(np.sqrt(x**2 + y**2)/(r*m*(1+np.sin(60*np.pi/180))))
    lam = np.arctan(y/x)
    return phi,lam
```
