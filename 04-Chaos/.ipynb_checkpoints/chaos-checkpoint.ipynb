```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
# Definition of Terms #
a = 0.25
b = 4.0
f = 8.0
g = 1.0
x01 = 2.5
x02 = 2.49
y0 = 1.0
z0 = 0.0

t0 = 0
tend = int(120*(24/4))
dtn = 4/(24*5)
```

```python
'''
dxdt = -y(t)**2 -z(t)**2 -(a*x(t)) + (a*f)
dydt = x*y - b*x*z - y + g
dzdt = b*x*y + x*z - z

x(t+dt) = x(t) + dt*dxdt(t)
ect....
'''
def x(y,z,aleph,dt):
    return aleph + dt*(- y**2 - z**2 - (a*aleph) + (a*f))
def y(x,z,aleph,dt):
    return aleph + dt*(x*aleph - b*x*z - aleph + g)
def z(y,x,aleph,dt):
    return aleph + dt*(b*x*y + x*aleph - aleph)
```

```python
print(x(y0,z0,x01,1))
```

```python
x1 = []
y1 = []
z1 = []
x2 = []
y2 = []
z2 = []

x1.append(x01)
y1.append(y0)
z1.append(z0)
x2.append(x02)
y2.append(y0)
z2.append(z0)
print(x1[0])
print(dtn)
```

```python
for i in range(1,tend+1):
    x1.append(x(y1[i-1],z1[i-1],x1[i-1],dtn))
    x2.append(x(y2[i-1],z2[i-1],x2[i-1],dtn))
    y1.append(y(x1[i-1],z1[i-1],y1[i-1],dtn))
    y2.append(y(x2[i-1],z2[i-1],y2[i-1],dtn))
    z1.append(z(y1[i-1],x1[i-1],z1[i-1],dtn))
    z2.append(z(y2[i-1],x2[i-1],z2[i-1],dtn))

print(z2[-1])
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1,layout="constrained")
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312,sharex=ax1)
ax3 = fig.add_subplot(313,sharex=ax1)

ax1.set_xlim(0,len(x1))
ax1.set_ylim(min(x2),max(x2))
ax2.set_ylim(min(y2),max(y2))
ax3.set_ylim(min(z2),max(z2))

ax1.plot(range(len(x1)),x1,color='r',label="x0 = 2.50")
ax1.plot(range(len(x1)),x2,color='black',label="x0 = 2.49")
ax2.plot(range(len(x1)),y1,color='r',label="x0 = 2.50")
ax2.plot(range(len(x1)),y2,color='black',label="x0 = 2.49")
ax3.plot(range(len(x1)),z1,color='r',label="x0 = 2.50")
ax3.plot(range(len(x1)),z2,color='black',label="x0 = 2.49")

ax1.legend()
ax2.legend()
ax3.legend()

plt.show()
```
